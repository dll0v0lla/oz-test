[문제1]

list와 tuple의 주된 차이점 은 무엇인가요?

리스트는 변경 가능하고, 튜플은 불변합니다.
리스트는 대괄호([])를 사용하고, 튜플은 소괄호(())를 사용합니다.
리스트는 변경 가능성과 동적 크기로 인해 주로 수정이 필요한 데이터에 사용됩니다. 튜플은 변경이 필요 없는 데이터나 해시 키와 같이 변경되지 않아야 하는 데이터에 사용됩니다.

[문제2]

Python의 set에 대해 설명해 주세요

집합(set)은 고유한 요소로 구성된 변경 가능한 컬렉션입니다.
중복을 허용하지 않으며, 순서가 없습니다.
중괄호({})를 사용하여 생성하고, 각 요소는 쉼표(,)로 구분됩니다.
집합은 합집합, 교집합, 차집합 등의 집합 연산을 제공하여 유용하게 사용됩니다.

[문제3]

Python의 list는 연결리스트(linked list) 인가요?

아닙니다. 리스트는 동적 배열(dynamic array)로 구현됩니다.
동적 배열은 메모리에 연속적으로 배치된 요소들로 이루어져 있어 인덱스를 사용하여 빠르게 요소에 접근할 수 있습니다.

[문제4]

딕셔너리 자료형에 대해 설명해주세요

딕셔너리는 키-값 쌍을 저장하는 해시 테이블 기반의 컬렉션입니다. 각 요소는 고유한 키(key)로 인덱싱되어 있으며, 값(value)은 중복될 수 있습니다.
중괄호({})를 사용하여 생성하고, 각 키-값 쌍은 쉼표(,)로 구분됩니다.
키를 사용하여 값을 검색하고 수정할 수 있습니다. 키는 불변 타입이어야 합니다.
